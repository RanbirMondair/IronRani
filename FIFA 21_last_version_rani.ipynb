```python
# to convert string data of value and wages columns to float value

def str2float(euros):
    # convert M in value column to millions
    if euros[-1] == 'M':
        return float(euros[1:-1])*1000000
    elif euros[-1] == 'K':
        return float(euros[1:-1])*1000
    else:
        return float(euros[1:])
    
# use lambda function to apply UF str2float() for each players market value
new_dataf['value'] = new_dataf['value'].apply(lambda x: str2float(x))

# use lambda function to apply UF str2float() for each players current wage 
new_dataf['wage'] = new_dataf['wage'].apply(lambda x: str2float(x))
```

```python
# choose all the columns we need
ml_cols =          [
                   'short_passing', 
                   'fk_accuracy',
                   'long_passing',
                   'ball_control',
                    'sprint_speed',
                   'shot_power', 
                   'long_shots',
                   'standing_tackle', 
                    'ova',
                   ]
```

```python
df_ml = pd.DataFrame(data=ml_players, columns=ml_cols)
```

```python
# check the data
df_ml.info()
```

```python
df_ml.isnull().any()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

```

```python
# Train test split
y = df_ml['ball_control']
X = df_ml[[
                   'short_passing', 
                   'fk_accuracy',
                   'long_passing', 
                    'sprint_speed',
                   'shot_power', 
                   'standing_tackle',
                   'long_shots',
                    'ova',
                   ]]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=101)
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

```python
print('Coefficients:', lm.coef_)
```

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test, predictions)
plt.xlabel('Y Test')
plt.ylabel('ball_control')
```

```python
# Evaluate the data
from sklearn import metrics
print("Mean absolute error", metrics.mean_absolute_error(y_test, predictions))
print("Mean squared error", metrics.mean_squared_error(y_test, predictions))
print("Root mean square error", np.sqrt(metrics.mean_squared_error(y_test, predictions)))
#  the results seems perfect
```

```python
sns.distplot((y_test - predictions), bins=50)
```

```python
# Get the effect of every parameter
coeffecients = pd.DataFrame(lm.coef_, X.columns)
coeffecients.columns = ['Coeffecient']
coeffecients
```
